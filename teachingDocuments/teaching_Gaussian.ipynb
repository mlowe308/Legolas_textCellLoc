```python
# Figure out how I can write text in Jupyter notebook. Maybe called "markdown."
# https://stackoverflow.com/questions/14873203/plotting-of-1-dimensional-gaussian-distribution-function
# Comments say the following:
# Your gaussian PDF is wrong - you need to scale by (\sqrt(2\pi)\sigma)^(-1). Additionally, x*x is much faster than pow(x, 2).
# shouldn't the x-axis be from -3 to 3?
# The old X values where counters, not the values of X (ie a mistake). (Explain)

from matplotlib import pyplot as mp
import numpy as np


def gaussian(x, mu, sig):
    return (
        1.0 / (np.sqrt(2.0 * np.pi) * sig) * np.exp(-np.power((x - mu) / sig, 2.0) / 2)
    )


x_values = np.linspace(-3, 3, 120)
for mu, sig in [(-1, 1), (0, 2), (2, 3)]:
    mp.plot(x_values, gaussian(x_values, mu, sig))

mp.show()
```

```python
# Example 1
# import the random module 
import random

# determining the values of the parameters 
mu = 100
sigma = 50

# using the gauss() method 
print(random.gauss(mu, sigma))
```

```python
# Example 2: We can generate the number multiple times and plot a graph to observe the gaussian distribution.
# import the required libraries  

# Make sure you know how to plot this graph by hand. Generate a few values and plot the values.

# In the computer graph, look at the mean and the values of the data points about the mean.
# How do I label the axes?

import random  
import matplotlib.pyplot as plt  
    
# store the random numbers in a   
# list  
nums = []  
mu = 100
sigma = 50
    
for i in range(100):  
    temp = random.gauss(mu, sigma) 
    nums.append(temp)  
        
# plotting a graph  
plt.plot(nums)  
plt.show() 
```

```python
# Example 3: We can create a histogram to observe the density of the gaussian distribution.

# import the required libraries

# Make sure you understand how to plot a histogram by hand. Generate a few values,
# create bins, and fill the bins.

import random  
import matplotlib.pyplot as plt  
    
# store the random numbers in a list  
nums = []  
mu = 100
sigma = 50
    
for i in range(10000):  
    temp = random.gauss(mu, sigma)  
    nums.append(temp)  
        
# plotting a graph  
plt.hist(nums, bins = 200)  
plt.show() 
```
